```python
import warnings
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from scipy.integrate import odeint
from scipy.constants import h, c, k
from math import *
from scipy.integrate import *
#constants
H = 71.0                            # HubbleConstant= 71 km/s/Mpc
Mpc= 3.085677581e+19                #kms
km = 1.0
Gyr=3.1536e16                       #seconds
H0 = (H * Gyr* km / Mpc)

t = np.arange(-2,4,0.01)
#equations 
#for case 1
a_1 = t
#for case 2 
p = (2*H0)**(0.5)
a_2 = p*t**(0.5)
#for case 3
def dadt(t,a):
    omega_m   = 1
    omega_lam = 0
    omega_T = omega_m + omega_lam
    i = H0*np.sqrt((omega_m/a)+(1-omega_T)+(omega_lam*a**2))
    return i
a0 = 1
sol = odeint(dadt,y0=a0,t=t,tfirst = True )
#plot
plt.plot(t,a_1+1, 'r-') 
plt.plot(t,sol, 'b-')
plt.xlabel(r' $t-t_0$ ')
plt.ylabel(r' $\frac{a(t)}{a_0}$ ')
plt.text(2, 2.6, r'$(0,0)$')

plt.plot(t,a_2+1, 'g-')


plt.grid(True)
plt.show()
warnings.filterwarnings("ignore",category=RuntimeWarning)
```
