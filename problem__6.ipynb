```python

import numpy as np
import matplotlib.pyplot as plt
## Constants
H = 71.0## HubbleConstant= 71 km/s/Mpc
Mpc= 3.085677581e+19#kms
km = 1.0
Gyr=3.1536e16   #seconds
H0 = (H * Gyr* km / Mpc)  
def atheta(oz, thetaa):
    eq = (1./2.) * (oz / (oz - 1.)) * (1. - np.cos(thetaa))
    return eq

def ttheta(oz, thetaa):
    eq = (1. / (2. * H)) * (oz / ((oz - 1.)**(3. / 2.))) * (thetaa - np.sin(thetaa)) / (km * Gyr / Mpc)
    return eq



thetaa = np.linspace(0, 2 * np.pi, num=361)  # Use degrees for angles
athetaa = []
tthetaa = []
closed = [1.1, 1.2, 1.5, 2.0]
agesclosed = [9.01115084854, 8.84833006354, 8.4238579855, 7.86623214832]

for i in range(len(closed)):
    athetaa.append(atheta(closed[i], thetaa))
    tthetaa.append(ttheta(closed[i], thetaa))
    plt.plot(tthetaa[i], athetaa[i])

for i in range(len(agesclosed)):
    plt.scatter(agesclosed[i], 1.0, 20, c='b', marker='o')
closedlegend= ['$\Omega_{M}=1.1$','$\Omega_{M}=1.2$','$\Omega_{M}=1.5$','$\Omega_{M}=2.0$']
plt.xlim(-10, 1510.0)
plt.ylim(0.0, 15.0)
plt.xlabel('t(Gyr)')
plt.ylabel('a(t)')
plt.legend(closedlegend)  
plt.grid(True)
plt.show()
```
